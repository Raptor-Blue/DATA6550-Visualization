### bad visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data/train.csv")
```

```python

survival_by_class = df.groupby("Pclass")["Survived"].mean()

plt.figure(figsize=(7,4))
plt.bar(["1st", "2nd", "3rd"], survival_by_class.values)


plt.ylim(0.40, 0.70)

plt.title("Extreme Survival Differences Between Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Data/train.csv")


df_bad = df[df["Age"].between(15, 40)].copy()

# Split survivors vs non-survivors
survivors = df_bad[df_bad["Survived"] == 1]["Age"].dropna()
nonsurvivors = df_bad[df_bad["Survived"] == 0]["Age"].dropna()

plt.figure(figsize=(7,4))


bins = [15, 18, 22, 26, 30, 34, 40]

plt.hist(survivors, bins=bins, alpha=0.9, label="Survived")
plt.hist(nonsurvivors, bins=bins, alpha=0.4, label="Did Not Survive")

plt.title("Survival Strongly Concentrated in Young Adults (Misleading View)")
plt.xlabel("Age")
plt.ylabel("Number of Passengers (Counts)")
plt.legend()
plt.show()
```
