```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Data/train.csv")


df_bad = df[df["Age"].between(15, 40)].copy()

# Split survivors vs non-survivors
survivors = df_bad[df_bad["Survived"] == 1]["Age"].dropna()
nonsurvivors = df_bad[df_bad["Survived"] == 0]["Age"].dropna()

plt.figure(figsize=(7,4))


bins = [15, 18, 22, 26, 30, 34, 40]

plt.hist(survivors, bins=bins, alpha=0.9, label="Survived")
plt.hist(nonsurvivors, bins=bins, alpha=0.4, label="Did Not Survive")

plt.title("Survival Strongly Concentrated in Young Adults (Misleading View)")
plt.xlabel("Age")
plt.ylabel("Number of Passengers (Counts)")
plt.legend()
plt.show()
```

```python
# Load data
df = pd.read_csv("Data/train.csv")
survival_by_class = df.groupby("Pclass")["Survived"].mean()

plt.figure(figsize=(7,4))
plt.bar(["1st", "2nd", "3rd"], survival_by_class.values)

plt.ylim(0, 1)  # FULL SCALE (ethical)
plt.title("Survival Rate by Passenger Class (Titanic)")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")

for i, v in enumerate(survival_by_class.values):
    plt.text(i, v + 0.02, f"{v:.2f}", ha="center")

plt.show()
```

```python
# Survival rate by gender
survival_by_gender = df.groupby("Sex")["Survived"].mean()

# Plot (ethical, clear)
plt.figure(figsize=(7,4))
plt.bar(survival_by_gender.index, survival_by_gender.values)

plt.ylim(0, 1)  # FULL SCALE (ethical)
plt.title("Survival Rate by Gender (Titanic)")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")

# Add value labels
for i, v in enumerate(survival_by_gender.values):
    plt.text(i, v + 0.02, f"{v:.2f}", ha="center")

plt.show()
```

```python
# Load dataset
df = pd.read_csv("Data/train.csv")
df = df.dropna(subset=["Age"])

# Separate survivors and non-survivors
survived = df[df["Survived"] == 1]
not_survived = df[df["Survived"] == 0]

plt.figure(figsize=(7,5))


plt.scatter(not_survived["Age"], not_survived["Fare"], alpha=0.5, label="Did Not Survive")
plt.scatter(survived["Age"], survived["Fare"], alpha=0.5, label="Survived")

plt.title("Age vs Fare Distribution (Titanic)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend()

plt.show()
```

```python


# Calculate survival rate by class
survival_by_class = df.groupby("Pclass")["Survived"].mean()

# -------------------------------
# GOOD VISUALIZATION
# -------------------------------
plt.figure(figsize=(7,4))
plt.bar(["1st", "2nd", "3rd"], survival_by_class.values)

plt.ylim(0, 1)  # Ethical: full scale
plt.title("GOOD: Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")

for i, v in enumerate(survival_by_class.values):
    plt.text(i, v + 0.02, f"{v:.2f}", ha="center")

plt.show()


# -------------------------------
# BAD VISUALIZATION
# -------------------------------
plt.figure(figsize=(7,4))
plt.bar(["1st", "2nd", "3rd"], survival_by_class.values)

plt.ylim(0.4, 0.7)  # Misleading: truncated axis
plt.title("BAD: Extreme Survival Differences Between Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")

for i, v in enumerate(survival_by_class.values):
    plt.text(i, v + 0.005, f"{v:.2f}", ha="center")

plt.show()
```

```python
# Calculate survival rate by passenger class
survival_by_class = df.groupby("Pclass")["Survived"].mean()

plt.figure(figsize=(6,6))

plt.pie(
    survival_by_class.values,
    labels=["1st Class", "2nd Class", "3rd Class"],
    autopct='%1.1f%%'
)

plt.title("GOOD: Survival Rate Distribution by Passenger Class")
plt.show()


plt.figure(figsize=(10,3))  # Wide shape exaggerates differences

plt.bar(
    ["1st Class", "2nd Class", "3rd Class"],
    survival_by_class.values
)

plt.ylim(0.4, 0.7)  # Truncated axis (misleading)

plt.title("BAD: Huge Survival Advantage for Elite Passengers")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")

plt.show()
```
